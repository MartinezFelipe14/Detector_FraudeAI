```python
X = df.drop('Class', axis=1).copy()

y = df['Class'].copy()

y.value_counts()

sns.countplot(x=y)
plt.title('DistribuiÃ§Ã£o das Classes')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.2, stratify=y)

# plotando a distribuiÃ§Ã£o das classes
```

```python
from xgboost import XGBClassifier
from imblearn.pipeline import Pipeline as imbPipeline
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import MinMaxScaler, StandardScaler

pipe = imbPipeline(steps=[('OVERSAMPLING', SMOTE()),
                          ('MinMaxScaler', MinMaxScaler()),
                          ('StandardScaler', StandardScaler()),
                          ('XGBClassifier', XGBClassifier(learning_rate=0.09518018573985505, min_child_weight=10,
                                                          max_depth=10, colsample_bytree=0.5,
                                                          gamma=0, scale_pos_weight=1, n_estimators=50))])

'''
learning_rate = params[0]
min_child_weight = params[1]
max_depth = params[2]
colsample_bytree = params[3]
gamma = params[4]
scale_pos_weight = params[5]
#[0.09518018573985505, 10, 10, 0.5, 0, 1]

'''


```

```python


from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
import numpy as np

cv = StratifiedKFold(n_splits=5, shuffle=True)

scores = []

# Itere sobre cada dobra do cross-validation
for split_treino, split_teste in cv.split(X, y):
    X_split_treino, X_split_teste = X.iloc[split_treino], X.iloc[split_teste]
    y_split_treino, y_split_teste = y.iloc[split_treino], y.iloc[split_teste]

    # Treine a pipeline no conjunto de dados de treinamento
    pipe.fit(X_split_treino, y_split_treino)

    # FaÃ§a previsÃµes no conjunto de dados de teste
    y_pred = pipe.predict(X_split_teste)

    # Calcule a acurÃ¡cia
    score = accuracy_score(y_split_teste, y_pred)
    scores.append(score)

print("MÃ©dia de acurÃ¡cia:", np.mean(scores))
```

```python
pipe.fit(X_treino, y_treino)

# fazendo as previsÃµes e as atibruindo para um variÃ¡vel y_pred, tambÃ©m poderia ser chamada de y_hat
y_pred = pipe.predict(X_teste)

from sklearn.metrics import classification_report

print(classification_report(y_teste, y_pred))  

```

```python

from sklearn.metrics import confusion_matrix

cnf_matrix = confusion_matrix(y_teste, y_pred)

# plotando uma matriz de confusÃ£o simples
sns.heatmap(cnf_matrix, annot=True, cmap='Blues', fmt='d')
plt.xlabel('y_pred')
plt.ylabel('Dados Reais')
plt.title('Matriz de ConfusÃ£o')
plt.show()

```

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_pred_proba = pipe.predict_proba(X_teste)[:,1]
fpr, tpr, _ = roc_curve(y_teste, y_pred_proba)
auc = roc_auc_score(y_teste, y_pred)

plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Taxa de Falsos Positivos')
plt.ylabel('Taxa de Verdadeiros Positivos')
plt.title('Curva ROC')
plt.show()
```
