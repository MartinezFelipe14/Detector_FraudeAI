```python
# definindo os dados de entrada
X = df.drop('Class', axis=1).copy()

# definindo o target ou a classe a ser prevista
y = df['Class'].copy()

y.value_counts()

# plotando a distribuição das classes
sns.countplot(x=y)
plt.title('Distribuição das Classes')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
from xgboost import XGBClassifier
from imblearn.pipeline import Pipeline as imbPipeline
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import MinMaxScaler

# criando uma pipeline de processos com oversampling, por isso é usado a Pipeline do ImbLearn
# parâmetros do XGBClassifier definidos pela otimização bayesiana no fine tuning
pipe = imbPipeline(steps=[('OVERSAMPLING', SMOTE()), # SMOTE convergiu melhor que ADASYN 
                          ('MinMaxScaler', MinMaxScaler()),
                          ('XGBClassifier', XGBClassifier(learning_rate=0.09518018573985505, min_child_weight=10, 
                                                          max_depth=10, colsample_bytree=0.5,
                                                          gamma=0, scale_pos_weight=1, n_estimators=50))])

'''
learning_rate = params[0]
min_child_weight = params[1]
max_depth = params[2]
colsample_bytree = params[3]
gamma = params[4]
scale_pos_weight = params[5]
#[0.09518018573985505, 10, 10, 0.5, 0, 1]

'''


```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
import numpy as np

cv = StratifiedKFold(n_splits=5, shuffle=True)

scores = []

# Iterando sobre cada fold do cross-validation para evitar data leakage e não comprometer a validação
for split_treino, split_teste in cv.split(X, y):
    X_split_treino, X_split_teste = X.iloc[split_treino], X.iloc[split_teste]
    y_split_treino, y_split_teste = y.iloc[split_treino], y.iloc[split_teste]

    # Fazendo os processos da pipeline somente no conjunto de dados de treinamento
    pipe.fit(X_split_treino, y_split_treino)

    # Fazendo previsões no conjunto de dados de teste
    y_pred = pipe.predict(X_split_teste)

    # Calculando a acurácia
    score = accuracy_score(y_split_teste, y_pred)
    scores.append(score)

print("Média de acurácia:", np.mean(scores))
```

```python
pipe.fit(X_treino, y_treino)

# fazendo as previsões e as atibruindo para um variável y_pred, também poderia ser chamada de y_hat
y_pred = pipe.predict(X_teste)

from sklearn.metrics import classification_report

print(classification_report(y_teste, y_pred))  

```

```python

from sklearn.metrics import confusion_matrix

cnf_matrix = confusion_matrix(y_teste, y_pred)

# plotando uma matriz de confusão
sns.heatmap(cnf_matrix, annot=True, cmap='Blues', fmt='d')
plt.xlabel('y_pred')
plt.ylabel('Dados Reais')
plt.title('Matriz de Confusão')
plt.show()

```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# criando um gráfico de curva roc, quanto mais perto do pontilhado pior
y_pred_proba = pipe.predict_proba(X_teste)[:,1]
fpr, tpr, _ = roc_curve(y_teste, y_pred_proba)
auc = roc_auc_score(y_teste, y_pred)

plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Taxa de Falsos Positivos')
plt.ylabel('Taxa de Verdadeiros Positivos')
plt.title('Curva ROC')
plt.show()
```
